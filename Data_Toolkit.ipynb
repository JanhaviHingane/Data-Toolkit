Data Toolkit

Q: What is NumPy, and why is it widely used in Python?

A: NumPy is a Python library for numerical and scientific computing. It is widely used because it provides fast multidimensional arrays and supports vectorized mathematical operations, making computation much faster than Python lists.

Q: How does broadcasting work in NumPy?

A: Broadcasting allows NumPy to perform arithmetic operations on arrays of different shapes by automatically expanding the smaller array to match the larger one.

Q: What is a Pandas DataFrame?

A: A DataFrame is a 2D labeled data structure in Pandas that stores data in rows and columns like an Excel or SQL table.

Q: Explain the use of the groupby() method in Pandas.

A: groupby() groups data based on certain conditions and then applies aggregation functions such as sum, mean, count, etc., on each group.

Q: Why is Seaborn preferred for statistical visualizations?

A: Seaborn provides visually appealing plots with fewer lines of code and has built-in support for statistical visualizations such as heatmaps, distribution plots, and regression plots.

Q: What are the differences between NumPy arrays and Python lists?

A: NumPy arrays are faster, more memory-efficient, fixed in data type, and support vectorized operations; Python lists are slower, allow mixed data types, and do not support vectorization.

Q: What is a heatmap, and when should it be used?

A: A heatmap uses colors to represent values in a matrix. It should be used to show relationships or correlations between variables.

Q: What does the term “vectorized operation” mean in NumPy?

A: It means performing arithmetic on entire arrays at once without using loops, which significantly speeds up computation.

Q: How does Matplotlib differ from Plotly?

A: Matplotlib generates static (non-interactive) plots, while Plotly generates interactive plots suitable for dashboards and web visualizations.

Q: What is the significance of hierarchical indexing in Pandas?

A: It enables multiple index levels for rows or columns, making it easier to work with grouped or multidimensional data.

Q: What is the role of Seaborn’s pairplot() function?

A: pairplot() visualizes pairwise relationships between numerical variables using scatterplots and histograms.

Q: What is the purpose of the describe() function in Pandas?

A: describe() provides statistical summary information such as count, mean, standard deviation, min, max, and quartiles for numerical columns.

Q: Why is handling missing data important in Pandas?

A: Missing data can lead to incorrect calculations or biased results, so it must be filled or removed for accurate analysis.

 Q: What are the benefits of using Plotly for data visualization?

A: Plotly creates interactive visualizations, supports 3D charts, dashboards, and integrates easily with web applications.

Q: How does NumPy handle multidimensional arrays?

A: NumPy uses ndarray, which supports N-dimensional data storage with efficient indexing and operations along each axis.

Q: What is the role of Bokeh in data visualization?

A: Bokeh is used to create interactive, web-based visualizations and dashboards with JavaScript integration.

Q: Explain the difference between apply() and map() in Pandas.

A: map() works on Series and transforms each value; apply() works on both Series and DataFrames and applies functions row-wise or column-wise.

Q: What are some advanced features of NumPy?

A: Linear algebra operations, random number generation, Fourier transforms, broadcasting, and memory-efficient multidimensional storage.

Q: How does Pandas simplify time series analysis?

A: Pandas supports date indexing, resampling, shifting, rolling statistics, and frequency conversion, making time-based analysis easier.

Q: What is the role of a pivot table in Pandas?

A: Pivot tables summarize data by grouping and aggregating values—similar to Excel’s pivot table.

Q: Why is NumPy’s array slicing faster than Python’s list slicing?

A: NumPy slicing returns a view of the same memory without copying data, while list slicing creates a new list and copies elements.

Q: What are some common use cases for Seaborn?

A: Visualizing distributions, correlations, pairwise plots, regression plots, heatmaps, and categorical comparisons.
